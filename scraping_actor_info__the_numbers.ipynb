```python
def request_page(url):
#     request_headers = make_headers()
    request_headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:29.0) Gecko/20120101 Firefox/29.0'}

    try:
        page = requests.get(url, headers=request_headers)
    except html.ConnectionError:
        print 'bad connection'
        return None

    if page.status_code != 200:
        print 'WARNING: Bad request'
        return None

    print 'request good: 200'

    tree = html.fromstring(page.content)
    
    return tree
```
